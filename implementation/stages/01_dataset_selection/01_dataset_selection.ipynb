## Candidate research and decision

Fashionpedia is the sole data source because its masks and fine-grained attributes support the required component tasks. The official training split supplies model training and internal validation, the labelled validation split remains the locked final test, and unlabelled test images are qualitative-only. Training uses every relevant labelled instance, while a separate manual whitelist controls images shown in coursework outputs.

```python
policy_path = DATA_ROOT / 'dataset-policy.json'
fashionpedia_root = DATA_ROOT / 'raw' / 'fashionpedia'
annotation_root = fashionpedia_root / 'annotations'
required_files = [
    annotation_root / 'instances_attributes_train2020.json',
    annotation_root / 'instances_attributes_val2020.json',
]
missing = [str(path) for path in required_files if not path.exists()]
if missing:
    raise FileNotFoundError(f'Missing required Fashionpedia files: {missing}')

policy = load_policy(policy_path)
print(json.dumps({
    'decision': 'retain Fashionpedia with relevant-instance training and a safe showcase',
    'policy_version': policy['policy_version'],
    'derived_targets': policy['derived_targets'],
    'minimum_support': policy['minimum_support'],
}, indent=2))
```
